```
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
sentences = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?']
```

```
tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')
tokenizer.fit_on_texts(sentences)
print(tokenizer.word_index)
```

```
sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)
```

```
# we can choose to pad from beginning or end
# if a sequence is longer than maxlen, it will be truncated, we can choose to truncate from beginning or end
padded = pad_sequences(sequences, maxlen=6)
print(padded)
```

```
test_sentences = [
    'I really love my dog',
    'My dog loves my manatee'
]
```

```
test_sequences = tokenizer.texts_to_sequences(test_sentences)
print(test_sequences)
```

```
test_padded = pad_sequences(test_sequences, maxlen=10)
print(test_padded)
```
